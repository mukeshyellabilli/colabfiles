```python
#Define a class BankAccount with attributes account_number and balance. Write methods deposit and withdraw to update the balance.
class BankAccount:
  def bank_details(self,account_number,balance):
    self.account_number=account_number
    self.balance=balance
    print('account number is',account_number,'balance is',balance)
  def deposit(self):
    print('deposit amount')
    amount=int(input())
    self.balance=self.balance+amount
    print('updated balance is',self.balance)
  def withdraw(self):
    print('withdraw amount')
    amount=int(input())
    self.balance=self.balance-amount
    print('updated balance',self.balance)
obj=BankAccount()
obj.bank_details(9391831463,5000)
obj.deposit()
obj.withdraw()
```

```python
#Define a class Account with attributes account_number and balance. Create subclasses CheckingAccount and SavingsAccount, each with their own implementation of a withdraw method.
class Account:
  def account_info(self,account_number,balance):
    self.account_number=account_number
    self.balance=balance
    print('account number : ',account_number)
    print('balance : ',balance)
class CheckingAccount(Account):
  def withdraw(self,amount):
    self.amount=amount
    print('withdraw amount : ',amount)
    print('updated balance : ',self.balance-amount)
class SavingsAccount(Account):
  def withdraw(self,amount):
    self.amount=amount
    print('withdraw amount : ',amount)
    print('updated balance : ',self.balance-amount)
obj=CheckingAccount()
obj.account_info(9701721637,6000)
obj.withdraw(1500)
obj1=SavingsAccount()
obj1.account_info(9701721637,4500)
obj1.withdraw(2000)
```

```python
#Create a base class Shape with a method draw. Implement subclasses Circle and Rectangle, each with their own implementation of the draw method.
class Shape:
  def draw(self):
    print('drawn shape')
class Circle(Shape):
  def draw(self):
    print('circle drawn')
    super().draw()
class Rectangle(Shape):
  def draw(self):
    print('rectangle drawn')
    super().draw()
obj=Circle()
obj.draw()
obj1=Rectangle()
obj1.draw()
```
