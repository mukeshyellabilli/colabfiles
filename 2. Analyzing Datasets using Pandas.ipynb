## **6. .Loc**

### **It is used for Location based filtering**
### **Also called as Name based filtering**

     Syntax: df.loc[ start_row_num : stop_row_num : row_step_num , start_col_name : stop_col_name : col_step_num ]
* Row index values should always be positive (step can be +ve/-ve)
* Columns must be names as specified in dataset, column indexes are not accepted


```python
df.loc[0:10,'age':'gender']   # with only start and stop values
```

```python
df.loc[0:10:2,'age':'gender':2]  # including step values
```

```python
df.loc[-1:-10:-1,'age':'gender'] # Doesn't work as row accepts only postive index values
```

```python
df.loc[400:300:-1,'age':'gender'] #printing rows in reverse by using positive index values
```

```python
df.loc[400:300:-1,'gender':'age':-1] # printing rows and columns in reverse
```

## **7. .I - loc**

### **Used for index location based filtering**
### **Also called Number based filtering**
     Syntax: df.iloc[ start_row_num : stop_row_num : row_step_num , start_col_num : stop_col_num : col_step_num ]

* Row index values should always be positive (step can be +ve/-ve)
* Columns index values can be both +ve/-ve

```python
df.iloc[0:10,5:10]  # with only start and stop values
```

```python
df.iloc[0:10:2,5:10:2]  # including step values
```

```python
df.iloc[10:50:-5,10:5:-2] #printing rows in reverse by using Negative index values (doesn't work as always row values should be positive)
```

```python
df.iloc[50:10:-5,5:10:2] #printing rows in reverse by using positive index values
```

```python
df.iloc[50:10:-5,10:5:-2] # printing rows and columns in reverse (using positive column indexes)
```

```python
df.iloc[50:10:-5,-1:-5:-2] # printing rows and columns in reverse (using postive row and negative column indexes)
```

## **17. Group by**

### groupby is useful when we need to apply multiple aggregated functions for their respective columns and retreive based on a particular column

    Syntax: df.groupby(column_to_be_grouped(with/without condition)).agg({'column1':'function1','column2':'function2'....})

* agg function takes dictionary with column names as key and aggregated functions name as value respectively
* It applies the value i.e aggregated function name to that particular column name in key

       key:value ---> 'column':'function'

* we cannot use multiple aggregated functions for a single column as it makes dublicates keys in agg dictionary according to the syntax of groupby

* we need to a different syntax for multiple aggregated functions for a single column i.e

        Syntax: df.groupby(column_to_be_grouped(with/without condition)).agg(new_col1 = ('col','func1'), new_col2 = ('col','func2'))
                   

```python
# Without using groupby

df.loc[df['race'] == 'White','age'].mean()

# This can be used when we need to apply a single aggregated function
```

```python
# Using groupby with condition and only single aggregated function

df.groupby(df['race'] == 'White').agg({'age': 'mean'})

# output explanation:
# False ---> mean of age of rows where race is not White
# True ---> mean of age of rows where race is White
```

```python
# Using groupby without any condition, this will consider all unique values in specified column

df.groupby(df['race']).agg({'age': 'mean'})
```

```python
# Using groupby without any condition and multiple aggregated functions

df.groupby(df['race']).agg({'age': 'mean','capital-gain':'max','capital-loss':'max'})
```

```python
df.groupby('race').agg(
                     min_age = ('age','min'),
                     max_age = ('age','max'),
                     mean_age = ('age','mean'),
                     min_hpw = ('hours-per-week','min'),
                     max_hpw = ('hours-per-week','max'),
                     mean_hpw = ('hours-per-week','mean'),
                    )
```
